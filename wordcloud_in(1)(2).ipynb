```python
text_data = wiki_obama.summary  # Replace 'wiki_obama.summary' with the actual variable holding your text
```

```python
cloud = WordCloud(max_words=2000, min_font_size=20, width=1000, height=1000,
                  background_color="white",
                  relative_scaling=0.5, random_state=14).generate_from_text(text_data)
plt.figure(figsize=(8, 6))
plt.imshow(cloud)
plt.axis("off")
plt.show()
```
